```python
# CSCE 580/581 Quiz2 Q2 part a
# Written by Tyler Beetle
from collections import deque
import time

# intialize the start state and goal state here
initial_state = (3,3,1)
goal_state = (0,0,0)

# Establishes the parameters for the missionaries, cannibals, and boat
def is_valid(state):
    m, c, b = state
    return 0 <= m <= 3 and 0 <= c <= 3 and (m >= c or m == 0) and (3 - m >= 3 - c or 3 - m == 0)

# Iterates through the states to find the best next possible combination and then stores the next state 
def successors(state):
    m,c,b = state
    next_states = []
    
    for dm in range(0, 4):
        for dc in range(0, 4):
            if 1 <= dm +dc <= 2:
                if b == 1:
                    next_m = m - dm
                    next_c = c - dc
                    next_b = 0
                else:
                    next_m = m + dm
                    next_c = c + dc
                    next_b = 1
               
                next_state = (next_m, next_c, next_b)
                
                if is_valid(next_state) and is_valid((3 - next_m, 3 - next_c, next_b)):
                    next_states.append(next_state)
                
    return next_states

#Breadth first search that reiterates until the goal state is reached, at which time the execution time is recorded
def bfs():
    visited = set()
    queue = deque([(initial_state, [])])
    
    start_time = time.time()
    
    while queue:
        current_state, path = queue.popleft()
                  
        if current_state == goal_state:
            path.append(current_state)
            return path, time.time() - start_time
        if current_state not in visited:
                  visited.add(current_state)
                  for successor in successors(current_state):
                      queue.append((successor, path + [current_state]))
    
    return None, time.time() - start_time
                  
solution_path, execution_time = bfs()

#Prints out the result to the user 
if solution_path:
    print("Solution Path:")
    for i, state in enumerate(solution_path):
            print(f"Step {i+1}: {state}")
    print(f"Execution Time: {execution_time: .6f} seconds")
else:
    print("No solution found.")
```

```python
# CSCE 580/581 Quiz2 Q2 part b
# Written by Tyler Beetle
import heapq
import time

# intialize the start state and goal state here
initial_state = (3,3,1)
goal_state = (0,0,0)

# Establishes the parameters for the missionaries, cannibals, and boat
def is_valid(state):
    m, c, b = state
    return 0 <= m <= 3 and 0 <= c <= 3 and (m >= c or m == 0) and (3 - m >= 3 - c or 3 - m == 0)

# Iterates through the states to find the best next possible combination and then stores the next state 
def successors(state):
    m,c,b = state
    next_states = []
    
    for dm in range(0, 4):
        for dc in range(0, 4):
            if 1 <= dm +dc <= 2:
                if b == 1:
                    next_m = m - dm
                    next_c = c - dc
                    next_b = 0
                else:
                    next_m = m + dm
                    next_c = c + dc
                    next_b = 1
               
                next_state = (next_m, next_c, next_b)
                
                if is_valid(next_state) and is_valid((3 - next_m, 3 - next_c, next_b)):
                    next_states.append(next_state)
                
    return next_states

# Defines an admissible heuristic, the sum of the difference
def heuristic(state):
    return state[0] + state [1]

# Solves the problem using the A* search algorithm discussed in class
def astar():
    closed_set = set()
    open_set = [(heuristic(initial_state), 0, initial_state, [])]
    
    start_time = time.time()
    
    while open_set:
        _, g, current_state, path = heapq.heappop(open_set)
        
        if current_state not in closed_set:
            closed_set.add(current_state)
            for successor in successors(current_state):
                heapq.heappush(open_set, (g + heuristic(successor), g + 1, successor, path + [current_state]))
        
        if current_state == goal_state:
            path.append(current_state)
            return path, time.time() - start_time
                
    return None, time.time() - start_time

solution_path, execution_time = astar()

#Prints out the result to the user 
if solution_path:
    print("Solution Path:")
    for i, state in enumerate(solution_path):
            print(f"Step {i+1}: {state}")
    print(f"Execution Time: {execution_time: .6f} seconds")
else:
    print("No solution found.")
```
